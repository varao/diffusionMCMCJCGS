```r
library('ggplot2')
library('tidyverse')
library('reshape2')
library('coda')
```

```r
source('../sder.R')

hyp_sin <- 2   # 1 in the sine drift

if(hyp_sin == 1) { 
    source('../ea1_func.R')
    ea_func <- ea1_func
    drift <- drift1
 } else { 
    source('../ea2_func.R')
    ea_func <- ea2_func
    drift <- drift2
 }
```

```r
set.seed(1)
x0 <- 0; tEnd <- 20; nObs <- 9
sdv <- .5
sample_size = 15000
Prt = 50
step_size=.01
theta=1

dat <- data_gen(x0 = 0, t0 = 0, sd = sdv, tEnd = tEnd, nObs = nObs,drift=function(x) drift(x,theta))
tDat <- dat$tDat
xDat <- dat$xDat
set.seed(1)
```

```r
# Fearnhead
pValue <- rep(0,length(tDat))
ksDist <- rep(0,length(tDat))

time_spend1 <- system.time(rslt_fearn <- 
                      pmcmcFearn(reps = sample_size, P=Prt, x0=x0, t0=0, tEnd=tEnd, xDat=xDat, tDat=tDat, 
                                 func = ea_func, theta=1, sd = sdv))
rslt_fearn <- t(rslt_fearn[2:(nObs+1)])
print("time spend is")
print(time_spend1)
timeCost_frn <- time_spend1["elapsed"]
es_frn <- effectiveSize(t(rslt_fearn))
ess_frn <- es_frn / timeCost_frn
```

```r
# Euler-pPMCMC   
time_spend3 <- system.time(rslt_eul <- 
                          pmcmc_Eul(reps = sample_size, P=Prt, t0=0, stepSize=step_size, tEnd=tEnd, xDat=xDat, 
                          tDat=tDat, sdv = sdv, theta=1, func=drift))
rslt_eul <- rslt_eul$xObsMat
print("time spend is")
print(time_spend3)
timeCost_eul <- time_spend3["elapsed"]
es_eul <- effectiveSize(t(rslt_eul))
ess_eul <- es_eul / timeCost_eul
```

```r
# HMC
time_spend2 <- system.time(rslt_hmc <- sde_hmc(reps = sample_size, x0=x0, t0=0,
                                         tEnd = tEnd, xDat = xDat, tDat = tDat,
                                         func = ea_func,
                                         sdv = sdv,
                                         theta = 1, M=100, eps=.2, L=5,
                                         ))
print("time spend is")
print(time_spend2)
timeCost_hmc <- time_spend2["elapsed"]
hmcObsMat <- rslt_hmc$xObsMat
es_hmc <- effectiveSize(t(hmcObsMat))
ess_hmc <- es_hmc / timeCost_hmc
```

```r
dfm_cmp <- c(Prt, tEnd, nObs, timeCost_frn, median(es_frn), median(ess_frn), 
                              timeCost_hmc, median(es_hmc), median(ess_hmc), 
                              timeCost_eul, median(es_eul), median(ess_eul))
t(dfm_cmp)
```

```r
source('../sder.R')

hyp_sin <- 1   # 1 in the sine drift

if(hyp_sin == 1) { 
    source('../ea1_func.R')
    ea_func <- ea1_func
    drift <- drift1
 } else { 
    source('../ea2_func.R')
    ea_func <- ea2_func
    drift <- drift2
 }
```

```r
x0 <- 0; tEnd <- 20; nObs <- 9
sdv <- .5
sample_size = 15000
Prt = 50
step_size=.01
theta=1


set.seed(1)

dat <- data_gen(x0 = 0, t0 = 0, sd = sdv, tEnd = tEnd, nObs = nObs,drift=function(x) drift(x,theta))
tDat <- dat$tDat
xDat <- dat$xDat
set.seed(1)
```

```r
time_spend2 <- system.time(rslt_eul <- 
                          pmcmc_Eul(reps = sample_size, P=Prt, t0=0, stepSize=0.01, tEnd=tEnd, 
                                    xDat=xDat, tDat=tDat, sdv = sdv, theta=NULL, func=drift))
```

```r
time_spend1 <- system.time(rslt_hmc <- sde_hmc(reps = sample_size, x0=x0, t0=0,
                                         tEnd = tEnd, xDat = xDat, tDat = tDat,
                                         func = ea_func,
                                         sdv = sdv,
                                         theta = NULL, M=100, eps=.2, L=5,
                                         ))
```

```r
timeCost_hmc <- time_spend1["elapsed"]
timeCost_eul <- time_spend2["elapsed"]

############
      
hmc_theta <- rslt_hmc$theta
es_hmc    <- effectiveSize(hmc_theta)
ess_hmc   <- es_hmc / timeCost_hmc
      
es_eul    <- effectiveSize(rslt_eul$theta)
ess_eul   <- es_eul / timeCost_eul
    
Idx       <- seq(1, sample_size, by = 50)
ksRslt    <- ks.test(rslt_hmc$theta[Idx], rslt_eul$theta[Idx])
pValue    <- ksRslt$p.value
ksDist    <- ksRslt$statistic
    
dfm_cmp   <- tibble(p=Prt, T=tEnd, N=nObs, tHMC=timeCost_hmc, esHMC=es_hmc, essHMC=ess_hmc, 
                                           tEul=timeCost_eul, esEul=es_eul, essEul=ess_eul, 
                                           pVal=pValue, ks=ksDist)
dfm_cmp
```
